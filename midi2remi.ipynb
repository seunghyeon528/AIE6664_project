## 60 chord -> word

```python
import glob
from pathlib import Path
from tqdm import tqdm

midi_root_dir = "/home/nas3/DB/Music_Deep/pop_music_transformer_data" 
path = Path(midi_root_dir)
midi_paths = list(path.rglob('*.midi'))

def extract_events(input_path):
    note_items, tempo_items = utils.read_items(input_path)
    note_items = utils.quantize_items(note_items)
    max_time = note_items[-1].end
    
    chord_items = utils.extract_chords(note_items)
    items = chord_items + tempo_items + note_items

    groups = utils.group_items(items, max_time)
    events = utils.item2event(groups)
    return events
   
all_events = []
for path in tqdm(midi_paths):
    path = str(path)
    events = extract_events(path)
    all_events.append(events)
    
    
```

```python
len(all_events)
```

```python
chord_type_list = []
for events in tqdm(all_events):
    for event in events:
        if event.value == "N:N":
            print("nan")
            break
        else:
            if event.name == "Chord":
                if event.text not in chord_type_list:
                    chord_type_list.append(str(event.text))
print(chord_type_list)
len(chord_type_list)
        
```

```python
## chord idx dict
chord2idx = {x:i for i, x in enumerate(chord_type_list)}
print(chord2idx)
import pickle
with open('chord2idx_dict.pkl', 'wb') as f:
    pickle.dump(chord2idx , f, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
## check number of bars
cnt = 0
for events in tqdm(all_events):
    cnt = 0
    for event in events:
        if event.name == "Bar":
            cnt+=1
    print(cnt)
```

```python
## save all events
with open('all_events.pkl', 'wb') as f:
    pickle.dump(all_events , f, protocol=pickle.HIGHEST_PROTOCOL)
    
with open('all_events.pkl', 'rb') as handle:
    all_events = pickle.load(handle)
```
